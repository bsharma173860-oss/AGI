````python
# CELL 2: PERSISTENT UNIVERSE STATE
# Universes save to disk after every run
# Load back exactly where they left off

class UniverseState:
    """
    Saves and loads complete universe state.
    Universe never resets — always continues from last stop.
    Like a real universe — time only moves forward.
    """

    def __init__(self, name):
        self.name = name
        self.file = os.path.join(UNIVERSE_DIR, f'{name}_universe.json')

    def save(self, universe_data):
        universe_data['last_run'] = datetime.now().isoformat()
        with open(self.file, 'w') as f:
            json.dump(universe_data, f, indent=2)

    def load(self):
        if os.path.exists(self.file):
            with open(self.file) as f:
                return json.load(f)
        return None

    def exists(self):
        return os.path.exists(self.file)

    def age_hours(self):
        if not self.exists(): return 0
        data = self.load()
        last = datetime.fromisoformat(data.get('last_run', datetime.now().isoformat()))
        return (datetime.now() - last).total_seconds() / 3600

print('UniverseState class ready.')
print()

# Check existing universe states
for name in ['emotion', 'physics', 'memory', 'mirror']:
    state = UniverseState(name)
    if state.exists():
        data = state.load()
        hours = state.age_hours()
        print(f'Universe {name}: t={data.get("t",0)} last_run={data.get("last_run","never")[:19]} ({hours:.1f}h ago)')
    else:
        print(f'Universe {name}: not created yet')
````

````python
# CELL 3: DAILY EXPANSION — RUN THIS MANUALLY OR AUTOMATICALLY
# This is what runs every day
# Loads universe state, runs steps, saves back

def run_emotion_universe(steps=100):
    """Run Senator Emotion universe. Load, evolve, save."""
    state_mgr = UniverseState('emotion')
    data = state_mgr.load() or {
        't': 0, 'beings': {}, 'events': [], 'knowledge_gained': []
    }

    t = data['t']
    n_beings = 20
    EMOTIONS = ['joy','grief','love','fear','trust','anger','hope','peace']

    # Restore beings
    beings = {}
    if data['beings']:
        for bid, bdata in data['beings'].items():
            beings[int(bid)] = bdata
    else:
        np.random.seed(42)
        for i in range(n_beings):
            beings[i] = {
                'state': {e: np.random.uniform(0,0.5) for e in EMOTIONS},
                'energy': np.random.uniform(0.5, 1.0),
                'age': 0,
                'relationships': {},
            }
            beings[i]['state']['joy'] = np.random.uniform(0.3, 0.8)

    events = data.get('events', [])[-50:]
    knowledge = data.get('knowledge_gained', [])

    # Run steps
    for _ in range(steps):
        t += 1
        for bid in list(beings.keys()):
            b = beings[bid]
            b['age'] += 1

            # Random interaction
            other_ids = [x for x in beings.keys() if x != bid]
            if other_ids:
                oid = random.choice(other_ids)
                o = beings[oid]

                # Love spreads
                if b['state']['joy'] > 0.5:
                    b['state']['love'] = min(1.0, b['state']['love'] + 0.03)
                    o['state']['love'] = min(1.0, o['state']['love'] + 0.015)
                    rel_key = str(oid)
                    b['relationships'][rel_key] = min(1.0,
                        b['relationships'].get(rel_key, 0) + 0.05)

                # Fear spreads from anger
                if b['state']['anger'] > 0.6:
                    o['state']['fear'] = min(1.0, o['state']['fear'] + 0.08)

            # Aging — energy drops
            b['energy'] = max(0.0, b['energy'] - 0.001)

            # GRIEF EMERGES from love + energy loss (not programmed)
            if b['energy'] < 0.3 and b['state']['love'] > 0.3:
                b['state']['grief'] = min(1.0, b['state']['grief'] + 0.04)
                b['state']['joy'] = max(0.0, b['state']['joy'] - 0.02)
                events.append(f't={t} grief_emerges being_{bid} energy={b["energy"]:.2f} love={b["state"]["love"]:.2f}')

            # HOPE EMERGES from community (not programmed)
            close = sum(1 for s in b['relationships'].values() if s > 0.4)
            if close >= 3:
                b['state']['hope'] = min(1.0, b['state']['hope'] + 0.008)
                b['state']['peace'] = min(1.0, b['state']['peace'] + 0.005)

            # WISDOM EMERGES from survived grief (not programmed)
            if b['state']['grief'] > 0.5 and b['age'] > 100:
                b['state']['peace'] = min(1.0, b['state']['peace'] + 0.015)
                b['state']['grief'] = max(0.0, b['state']['grief'] - 0.008)
                if random.random() < 0.01:
                    events.append(f't={t} wisdom_emerges being_{bid} survived grief becomes peace')

    # What the senator learned today
    avg = {e: np.mean([b['state'][e] for b in beings.values()]) for e in EMOTIONS}
    dominant = max(avg.items(), key=lambda x: x[1])
    today_knowledge = (
        f'after {steps} steps dominant emotion is {dominant[0]} {dominant[1]:.2f} . '
        f'grief emerging in beings with low energy and high love . '
        f'hope building in beings with three or more close relationships . '
        + ' . '.join(events[-5:])
    )
    knowledge.append({'t': t, 'text': today_knowledge, 'date': datetime.now().isoformat()})

    # Save state
    state_mgr.save({'t': t, 'beings': beings, 'events': events[-100:],
                    'knowledge_gained': knowledge[-30:], 'avg_emotions': avg})
    return t, avg, events[-5:], today_knowledge


def run_memory_universe(steps=200):
    """Run Senator Memory universe. Civilizations rise and fall."""
    state_mgr = UniverseState('memory')
    data = state_mgr.load() or {'t':0,'civs':{},'events':[],'knowledge_pool':0.5,'knowledge_gained':[]}

    t = data['t']
    knowledge_pool = data.get('knowledge_pool', 0.5)
    events = data.get('events', [])[-50:]
    knowledge = data.get('knowledge_gained', [])

    CIV_NAMES = ['Vedic','Hebrew','Greek','Chinese','Egyptian','Mesopotamian','Roman','Persian']
    civs = data.get('civs', {})
    if not civs:
        np.random.seed(42)
        for i, name in enumerate(CIV_NAMES):
            civs[str(i)] = {'name':name,'complexity':np.random.uniform(0.1,0.5),
                            'knowledge':np.random.uniform(0.1,0.4),'corruption':0.0,
                            'external_pressure':0.0,'phase':'rising','age':0}

    for _ in range(steps):
        t += 1
        new_civs = {}
        for cid, c in civs.items():
            c['age'] += 1
            if c['phase'] == 'rising':
                c['knowledge'] = min(1.0, c['knowledge'] + 0.008)
                c['complexity'] = min(1.0, c['complexity'] + 0.006)
                knowledge_pool += 0.0005
                if c['complexity'] > 0.75:
                    c['phase'] = 'peak'
                    events.append(f't={t} {c["name"]} reaches peak complexity')
            elif c['phase'] == 'peak':
                c['corruption'] = min(1.0, c['corruption'] + 0.004)
                c['external_pressure'] = min(1.0, c['external_pressure'] + 0.003)
                if c['corruption'] + c['external_pressure'] > 0.7:
                    c['phase'] = 'declining'
                    events.append(f't={t} {c["name"]} begins decline corruption={c["corruption"]:.2f}')
            elif c['phase'] == 'declining':
                c['complexity'] = max(0.0, c['complexity'] - 0.01)
                knowledge_pool += c['knowledge'] * 0.003
                c['knowledge'] = max(0.0, c['knowledge'] - 0.002)
                if c['complexity'] < 0.05:
                    c['phase'] = 'fallen'
                    events.append(f't={t} {c["name"]} civilization falls knowledge preserved pool={knowledge_pool:.3f}')
                    new_id = str(max(int(k) for k in civs.keys()) + 1)
                    new_civs[new_id] = {'name':f'Post-{c["name"]}',
                        'complexity':0.12,'knowledge':min(0.4,knowledge_pool*0.25),
                        'corruption':0.0,'external_pressure':0.0,'phase':'rising','age':0}
                    events.append(f't={t} new civilization Post-{c["name"]} rises knowledge={new_civs[new_id]["knowledge"]:.2f}')
        civs.update(new_civs)

    phases = defaultdict(list)
    for c in civs.values(): phases[c['phase']].append(c['name'])
    today_knowledge = (
        f'civilizations rise and fall at t {t} . '
        f'knowledge pool {knowledge_pool:.3f} grows as civilizations fall . '
        f'rising {phases.get("rising",[])} . fallen {phases.get("fallen",[])} . '
        + ' . '.join(events[-4:])
    )
    knowledge.append({'t':t,'text':today_knowledge,'date':datetime.now().isoformat()})

    state_mgr.save({'t':t,'civs':civs,'events':events[-100:],'knowledge_pool':knowledge_pool,
                    'knowledge_gained':knowledge[-30:]})
    return t, dict(phases), knowledge_pool, events[-5:]


def run_mirror_universe(steps=50):
    """Run Mirror Universe — Bharat building AGI."""
    state_mgr = UniverseState('mirror')
    data = state_mgr.load() or {'t':0,'bharat':{},'events':[],'knowledge_gained':[]}

    t = data['t']
    events = data.get('events', [])[-50:]
    knowledge = data.get('knowledge_gained', [])

    FOCUSES = ['parliament_agi','paper_writing','benchmarking','senator_universes','arxiv_submission','funding']
    b = data.get('bharat', {
        'focus':'parliament_agi','progress':0.3,'insight':0.7,'energy':0.8,
        'excitement':0.9,'confidence':0.75,'frustration':0.1,
        'theorems':8,'senators':6,'age':0
    })

    for _ in range(steps):
        t += 1
        b['age'] += 1
        b['progress'] = min(1.0, b['progress'] + b['insight'] * b['energy'] * 0.015)
        b['frustration'] = max(0.0, b['frustration'] - 0.008)

        if np.random.random() < 0.08:
            obstacle = np.random.choice(['api_credits','disk_space','model_quality','endorser_search'])
            b['frustration'] = min(1.0, b['frustration'] + 0.12)
            b['energy'] = max(0.2, b['energy'] - 0.04)
            if b['confidence'] > 0.6:
                b['insight'] = min(1.0, b['insight'] + 0.04)
                events.append(f't={t} obstacle:{obstacle} insight grows to {b["insight"]:.2f}')

        if b['progress'] >= 1.0:
            b['excitement'] = min(1.0, b['excitement'] + 0.15)
            b['confidence'] = min(1.0, b['confidence'] + 0.04)
            b['frustration'] = max(0.0, b['frustration'] - 0.25)
            idx = FOCUSES.index(b['focus']) if b['focus'] in FOCUSES else 0
            b['focus'] = FOCUSES[(idx+1) % len(FOCUSES)]
            b['progress'] = 0.0
            b['theorems'] += np.random.randint(0,2)
            events.append(f't={t} breakthrough focus now {b["focus"]} theorems={b["theorems"]}')

        if b['energy'] < 0.4:
            b['energy'] = min(1.0, b['energy'] + 0.06)

    today_knowledge = (
        f'bharat at t {t} focus {b["focus"]} progress {b["progress"]:.2f} . '
        f'insight {b["insight"]:.2f} excitement {b["excitement"]:.2f} . '
        f'theorems proven {b["theorems"]} senators built {b["senators"]} . '
        + ' . '.join(events[-3:])
    )
    knowledge.append({'t':t,'text':today_knowledge,'date':datetime.now().isoformat()})

    state_mgr.save({'t':t,'bharat':b,'events':events[-100:],'knowledge_gained':knowledge[-30:]})
    return t, b, events[-5:]


print('All universe functions ready.')
print('Call run_emotion_universe(), run_memory_universe(), run_mirror_universe()')
````

````python
# CELL 7: CHECK UNIVERSE STATE ANYTIME
def parliament_status():
    """See current state of all universes at any time."""
    print('PARLIAMENT UNIVERSE STATUS')
    print(f'Time: {datetime.now().strftime("%Y-%m-%d %H:%M")}')
    print('='*55)
    print()

    for name in ['emotion', 'physics', 'memory', 'mirror']:
        state = UniverseState(name)
        if state.exists():
            data = state.load()
            hours = state.age_hours()
            t = data.get('t', 0)
            last = data.get('last_run', 'never')[:19]
            print(f'Universe {name.upper()}:')
            print(f'  Time: t={t} | Last run: {last} ({hours:.1f}h ago)')

            if name == 'emotion' and data.get('avg_emotions'):
                avg = data['avg_emotions']
                dom = max(avg.items(), key=lambda x: x[1])
                print(f'  Dominant emotion: {dom[0]} ({dom[1]:.3f})')

            if name == 'memory' and data.get('knowledge_pool'):
                print(f'  Knowledge pool: {data["knowledge_pool"]:.3f}')

            if name == 'mirror' and data.get('bharat'):
                b = data['bharat']
                print(f'  Bharat: focus={b.get("focus")} theorems={b.get("theorems",8)}')

            if data.get('knowledge_gained'):
                latest = data['knowledge_gained'][-1]
                print(f'  Latest learning: {latest["text"][:80]}...')
            print()
        else:
            print(f'Universe {name.upper()}: not created yet')
            print()

parliament_status()
````
